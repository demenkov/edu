```python
import os
from PIL import Image, ImageDraw

# img1.jpg — цветное 800x600 (RGB)
if not os.path.exists("img1.jpg"):
    im = Image.new("RGB", (800, 600), (70, 130, 180))
    d = ImageDraw.Draw(im)
    for x in range(0, 800, 80):
        for y in range(0, 600, 80):
            if (x // 80 + y // 80) % 2 == 0:
                d.rectangle([x, y, x + 80, y + 80], fill=(210, 180, 140))
    d.ellipse([300, 200, 500, 400], fill=(220, 60, 60))
    im.save("img1.jpg", "JPEG")
    print("Создан img1.jpg")

# img2.gif — цветное 400x300
if not os.path.exists("img2.gif"):
    im = Image.new("RGB", (400, 300), (60, 180, 75))
    d = ImageDraw.Draw(im)
    d.rectangle([50, 50, 350, 250], outline=(255, 255, 255), width=5)
    d.line([50, 50, 350, 250], fill=(255, 255, 0), width=3)
    im.save("img2.gif", "GIF")
    print("Создан img2.gif")
```

```python
from PIL import Image

img = Image.open("img1.jpg")
# В Jupyter вместо img.show() выводим изображение в ячейку:
display(img)
```

```python
### 1.2. Открытие того же файла и перевод в оттенки серого (режим `L`)
```

```python
img = Image.open("img1.jpg")
img = img.convert("L")
print("Режим:", img.mode)
display(img)
```

```python
### 1.3. Передача файлового объекта, открытого в бинарном режиме `rb`
#Вместо пути к файлу в `Image.open()` можно передать сам файловый объект.
```

```python
# Открываем файл в бинарном режиме
f = open("img2.gif", "rb")
# Передаём объект файла
img = Image.open(f)
display(img)
f.close()  # закрываем файл
```

```python
### 1.4. Загрузка из строки байтов через `io.BytesIO`
#Предварительно читаем файл в бинарном режиме методом `.read()`.
```

```python
import io
from PIL import Image

with open("img2.gif", "rb") as f:
    data = f.read()              # содержимое файла как bytes

img = Image.open(io.BytesIO(data))
img = img.convert("L")
display(img)
```
